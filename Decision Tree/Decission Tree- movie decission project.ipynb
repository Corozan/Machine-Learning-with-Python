```python
import pandas as pd
```

```python
data = pd.read_csv('regressive.csv')
```

```python
data
```

```python
data.head()
```

```python
data.info() # to analyse the data
```

```python
data =pd.get_dummies(data,columns = ['3D_available','Genre']) #to tranfer one type of data set to another 
```

```python
data.info() #now we can 
```

```python
data
```

```python
 data['Time_taken'] = data['Time_taken'].fillna(data['Time_taken'].mean())
```

```python
data.info()
```

```python
# let collections as label and remaining all as the features 
```

```python
Y = data[['Collection']] # Labels
```

```python
Y
```

```python
# features
X = data.drop(['Collection'],axis=1) #drop the column Collections and take all other data , axis =1 is for to mention column
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size = 0.15)
```

```python
from sklearn import tree
model = tree.DecisionTreeRegressor(max_depth = 3) # max_depth is how depth should your tree be
```

```python
model.fit(xtrain,ytrain)
```

```python
ypredict = model.predict(xtest)
```

```python
ypredict
```

```python
model
```

```python
dot = tree.export_graphviz(model, out_file = None)
from IPython.display import Image
import pydotplus

graph = pydotplus.graph_from_dot_data(dot)
Image(graph.create_png())
```
